```python
import cv2
import easyocr
from matplotlib import pyplot as plt
import imutils
import numpy as np
```

```python
img = cv2.imread('car_img1.jpeg')

if img is None:
    print("Image not found. Check the file name or path.")
else:
    print("Image loaded successfully:", img.shape)
```

```python
img=cv2.resize(img,(640,480))
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray = cv2.bilateralFilter(gray, 11, 17, 17)
```

```python
plt.imshow(cv2.cvtColor(gray,cv2.COLOR_BGR2RGB))
```

```python
edged=cv2.Canny(gray,30,200)
```

```python
plt.imshow(cv2.cvtColor(edged,cv2.COLOR_BGR2RGB))
```

```python
nts = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(nts)
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)[:10]
screenCnt = None
```

```python
for c in cnts:
    peri=cv2.arcLength(c,True)
    approx=cv2.approxPolyDP(c,0.018*peri,True)
    if len(approx)==4:
        screenCnt=approx
        break
```

```python
screenCnt
```

```python
mask=np.zeros(gray.shape,np.uint8)
new_image=cv2.drawContours(mask,[screenCnt],0,255,-1)
new_image=cv2.bitwise_and(img,img,mask=mask)
```

```python
plt.imshow(cv2.cvtColor(new_image,cv2.COLOR_BGR2RGB))
```

```python
(x,y)=np.where(mask==255)
(x1,y1)=(np.min(x),np.min(y))
(x2,y2)=(np.max(x),np.max(y))
cropped_image=gray[x1:x2+1,y1:y2+1]
```

```python
plt.imshow(cv2.cvtColor(cropped_image,cv2.COLOR_BGR2RGB))
```

```python
reader=easyocr.Reader(['en'])
result=reader.readtext(cropped_image)
result
```
